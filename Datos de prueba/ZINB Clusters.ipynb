# Crear datos completos
1. 8 nubes de datos (clusters)
2. N√∫mero de genes: 10.000 
3. La probabilidad de la inflaci√≥n de ceros se selecciona de forma aleatoria entre 0 y 0.3
4. Los valores de lambda para las distribuciones de negative_binomial se seleccionan de forma aleatoria entre 1 y 10, dado que no queremos valores muy altos (hay pocos conteos altos en los datos de Single Cell)
5. El n√∫mero de datos de cada nube es un n√∫mero aleatorio entre 200 y 1000.

```python
def zero_inflated(samples, prob_zi):
    """
    A partir de una matriz de datos de negative_binomial, se ingresan ceros de forma aleatoria con probabilidad prob_zi
    """
    for i in range(samples.shape[0]):
        for j in range(samples.shape[1]):
            aleatorio = random.uniform(0, 1)
            if aleatorio <= prob_zi:
                samples[i,j] = 0

def create_cluster(dimension):
    """
    A partir de un arreglo d-dimensional de lambdas, un n√∫mero de datos y una probabilidad, se crea una nube de datos que siguen
    la distribuci√≥n ZINB. 
    """
    r = random.uniform(0.1, 1.5)
    sample_size = random.randint(500, 1500)
    prob = random.uniform(0.3, 1)

    samples = np.random.negative_binomial(r, prob, size=(sample_size, dimension))
    zero_inflated(samples, prob)

    return samples, r, sample_size, prob
```

```python
# Ocho clusters
dimension = 10000

# Primer cluster
total_samples, r, sample_size, prob =  create_cluster(dimension)

y = [1]*sample_size
total_rs = [r]
total_prob_zi = [prob]

# Clusters 2-8
for cluster in range(2,9):
    samples, r, sample_size, prob  = create_cluster(dimension)
    total_samples = np.vstack([total_samples, samples])
    y += [cluster]*sample_size 
    total_rs += [r]
    total_prob_zi += [prob]
```

```python
# Guardamos los datos
dic_clusters = {'X': total_samples, 'y': y, 'probability_ZI': total_prob_zi, 'total_rs': total_rs}

# with open('generados/zi_negative_binomial_dim10000.pkl', 'wb') as f:
#     pickle.dump(dic_clusters, f)
```

# Pasar a formato est√°ndar

```python
import pandas as pd
from scipy.sparse import csc_matrix
from scipy.io import mmwrite
```

```python
ruta_carpeta = 'generados/zi_negative_binomial_dim10000/'
sparse_X = csc_matrix(dic_clusters['X'])

# Verifica si la carpeta ya existe
if not os.path.exists(ruta_carpeta):
    os.makedirs(ruta_carpeta)
    os.makedirs(ruta_carpeta + 'hg19/')
```

```python
mmwrite(ruta_carpeta + 'hg19/matrix.mtx', sparse_X.T)

celulas = np.array(range(sparse_X.shape[0]))
c = pd.DataFrame(columns = ['c0'])
c['c0'] = celulas[1:]
c['c0'] = c['c0'].astype(str).apply(lambda x: 'c' + x)

genes = np.array(range(sparse_X.shape[1]))
g = pd.DataFrame(columns = ['g0', 'g00'])
g['g0'] = genes[1:]
g['g0'] = g['g0'].astype(str).apply(lambda x: 'g' + x)
g['g00'] = g.g0
```

```python
c.to_csv(ruta_carpeta+'hg19/barcodes.tsv', sep = '\t', index = None)
g.to_csv(ruta_carpeta+'hg19/genes.tsv', sep = '\t', index = None)
```
