# Poisson con inflación de ceros en una y dos dimensiones

```python
# Una dimensión 
samples = np.random.poisson([2], size=(10000, 1))

prob_zi = 0.3
for i in range(samples.shape[0]):
    for j in range(samples.shape[1]):
        aleatorio = random.uniform(0, 1)
        if aleatorio <= prob_zi:
            samples[i,j] = 0

plt.hist(samples, alpha = 0.5)
```

# Crear datos completos
1. 8 nubes de datos (clusters)
2. Número de genes: 10.000 
3. La probabilidad de la inflación de ceros se selecciona de forma aleatoria entre 0 y 0.3
4. Los valores de lambda para las distribuciones de poisson se seleccionan de forma aleatoria entre 1 y 10, dado que no queremos valores muy altos (hay pocos conteos altos en los datos de Single Cell)
5. El número de datos de cada nube es un número aleatorio entre 200 y 1000.

```python
def zero_inflated(samples, prob_zi):
    """
    A partir de una matriz de datos de Poisson, se ingresan ceros de forma aleatoria con probabilidad prob_zi
    """
    for i in range(samples.shape[0]):
        for j in range(samples.shape[1]):
            aleatorio = random.uniform(0, 1)
            if aleatorio <= prob_zi:
                samples[i,j] = 0

def create_cluster(dimension):
    """
    A partir de un arreglo d-dimensional de lambdas, un número de datos y una probabilidad, se crea una nube de datos que siguen
    la distribución ZINB. 
    """
    lambdas = [random.uniform(0.2, 2) for _ in range(dimension)]
    sample_size = random.randint(500, 1500)
    prob_zi = random.uniform(0, 0.3)

    samples = np.random.poisson(lambdas, size=(sample_size, len(lambdas)))
    zero_inflated(samples, prob_zi)

    return samples, lambdas, sample_size, prob_zi
```

```python
# Ocho clusters
dimension = 10000

# Primer cluster
total_samples, lambdas, sample_size, prob_zi =  create_cluster(dimension)

y = [1]*sample_size
total_lambdas = lambdas
total_prob_zi = [prob_zi]

# Clusters 2-8
for cluster in range(2,9):
    samples, lambdas, sample_size, prob_zi = create_cluster(dimension)
    total_samples = np.vstack([total_samples, samples])
    y += [cluster]*sample_size 
    total_lambdas += lambdas
    total_prob_zi += [prob_zi]
```

```python
# Guardamos los datos
dic_clusters = {'X': total_samples, 'y': y, 'lambdas_poisson': total_lambdas, 'probability_ZI': total_prob_zi}

# with open('generados/zi_poisson_dim10000.pkl', 'wb') as f:
#    pickle.dump(dic_clusters, f)
```

# Pasar a formato estándar

```python
import pandas as pd
from scipy.sparse import csc_matrix
from scipy.io import mmwrite
```

```python
ruta_carpeta = 'generados/zi_negative_poisson_dim10000/'
sparse_X = csc_matrix(dic_clusters['X'])

# Verifica si la carpeta ya existe
if not os.path.exists(ruta_carpeta):
    os.makedirs(ruta_carpeta)
    os.makedirs(ruta_carpeta + 'hg19/')
```

```python
mmwrite(ruta_carpeta + 'hg19/matrix.mtx', sparse_X.T)

celulas = np.array(range(sparse_X.shape[0]))
c = pd.DataFrame(columns = ['c0'])
c['c0'] = celulas[1:]
c['c0'] = c['c0'].astype(str).apply(lambda x: 'c' + x)

genes = np.array(range(sparse_X.shape[1]))
g = pd.DataFrame(columns = ['g0', 'g00'])
g['g0'] = genes[1:]
g['g0'] = g['g0'].astype(str).apply(lambda x: 'g' + x)
g['g00'] = g.g0
```

```python
c.to_csv(ruta_carpeta+'hg19/barcodes.tsv', sep = '\t', index = None)
g.to_csv(ruta_carpeta+'hg19/genes.tsv', sep = '\t', index = None)
```
